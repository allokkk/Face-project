```python
import cv2
import numpy as np
```

```python
#read and show video stream capture Image
#detect face show bonding box
#Flattern the largest face and save into numpy array
#Repeat the above for multiple people to generate training data



# initialise the camera
cap=cv2.VideoCapture(0)

#face detection using harcascade show bonding boxes

face_cascade=cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")

skip=0
#to strore the image in a array

face_data=[]
dataset_path="./data/"

file_name=input("enter your name")




while True:
    ret,frame=cap.read()
    
    
    if ret==False:
        continue
        
     #to conver the frme into gray frame
    gray_frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    
    
        
      
    faces=face_cascade.detectMultiScale(frame,1.3,5)    
    #print(faces)
    faces=sorted(faces,key=lambda f:f[2]*f[3])
    
    
      #pick the last face coz last face is having large area
    #drawing rectangle in faces
    #to iterate over the faces
    
    for face in faces[-1:]:
        x,y,w,h=face
        
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,255),2)
        
        
    
    #Exract (Crop requirred area):Region of interest
        offset=10
        face_section=frame[y-offset:y+h+offset,x-offset:x+w+offset]
    #to resise the face data
        face_section=cv2.resize(face_section,(100,100))
    
    
    #to store data on every tenth image
    skip+=1
    if (skip%10==0):
        face_data.append(face_section)
        print(len(face_data))
    
    
  
    
    cv2.imshow("Frame",frame)
    cv2.imshow("face section",face_section)
        
    
    
    
    #store every 10th face
    if(skip%10==0):
        #store the 10th face
        pass
    
    key_pressed=cv2.waitKey(1)&0xFF
    if key_pressed==ord('q'):
        break
        
#conver a facelist into numpy array

face_data=np.asarray(face_data)
print(face_data)
face_data=face_data.reshape((face_data.shape[0],-1))
print(face_data.shape)

#save this data into file system
np.save(dataset_path+file_name+'.npy',face_data)
print("data succesfully stored at"+dataset_path+file_name+'.npy')
        
cap.release()
cv2.destroyAllWindows()
```
